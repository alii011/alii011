```python
conditionlist = [
    (df["C1"]==0) & (df["C2"]==0) & (df["C3"]>=3) & (df["C4"]==0) & (df["C5"]==1),
    (df["C1"]==1) & (df["C2"]==1) & (df["C3"]==3) & (df["C4"]==1) & (df["C5"]==1),
    (df["C1"]==1) & (df["C2"]==1) & (df["C3"]==0) & (df["C4"]==1) & (df["C5"]==0),
    (df["C1"]==1) & (df["C2"]==1) & (df["C3"]<=2) & (df["C4"]==1) & (df["C5"]==0),
    (df["C3"]==5) & (df["C5"]==1),
    (df["C1"]==0) & (df["C2"]==1) & (df["C3"]<=2) & (df["C4"]==1),
    (df["C1"]==0) & (df["C2"]==1)  | (df["C1"]==1) & (df["C4"]==0) | (df["C2"]==1) & (df["C4"]==0) | (df["C4"]==1) & (df["C2"]==0),
    (df["C1"]==0) & (df["C3"]>3) | (df["C2"]==0) & (df["C3"]>3)| (df["C4"]==0) & (df["C3"]>3),
    (df["C1"]==1) & (df["C3"]>3) | (df["C2"]==1) & (df["C3"]>3)| (df["C4"]==1) & (df["C3"]>3),
    (df["C1"]==0) & (df["C3"]<=2) | (df["C2"]==0) & (df["C3"]<=2)| (df["C4"]==0) & (df["C3"]<=2),
    (df["C1"]==1) & (df["C3"]<=2) | (df["C2"]==1) & (df["C3"]<=2)| (df["C4"]==1) & (df["C3"]<=2), 
    (df["C5"]==0) & (df["C2"]==1)  | (df["C1"]==1) & (df["C5"]==0) | (df["C4"]==1) & (df["C5"]==0),
    (df["C5"]==0) & (df["C3"]>3),
    (df["C5"]==1) & (df["C3"]>3) ,
    (df["C5"]==0) & (df["C3"]<=2),
    (df["C5"]==1) & (df["C3"]<=2),
    ]
choicelist = ['100',random.randint(62,78), '0',random.randint(1,20),'80',random.randint(40,55),random.randint(50,80),random.randint(50,80),random.randint(50,65),random.randint(40,65),random.randint(30,45),random.randint(10,35),random.randint(50,70),random.randint(60,75),random.randint(20,45),random.randint(30,45)]
df['score'] = np.select(conditionlist, choicelist, default='Not Specified')

print(df)
```
